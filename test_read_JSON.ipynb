```python
import pandas as pandas

# read json file
df = pandas.read_json('data/puzzles_all.json', orient='index')

df.head(10)
```

```python
# calculate the average time to solve by puzzle type for puzzles that were solved
df_solved = df[df['solved'] == True]
df_solved.groupby('puzzle_type')[['time_to_solve', 'node_count']].agg(['mean', 'count'])
```
